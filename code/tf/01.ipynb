```python
import tensorflow as tf
import numpy as np
```

```python
x=np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
y=np.array([-40,  14, 32, 46, 59, 72, 100],dtype=float)
for i,c in enumerate(x):
    print("{} degress celsius = {} degrees Fahrenheit".format(c,y[i]))
```

```python
l0=tf.keras.layers.Dense(units=1,input_shape=(1,))
model=tf.keras.Sequential([l0])
model.compile(loss=tf.keras.losses.mse,optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history=model.fit(x,y,epochs=500,verbose=False)
print("Finished training the model")
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
print(model.predict([100.0]))
```

```python
print("These are the layer variables: {}".format(l0.get_weights()))
```

```python
l0=tf.keras.layers.Dense(units=4,input_shape=[1])
l1=tf.keras.layers.Dense(units=4)
l2=tf.keras.layers.Dense(units=1)
model=tf.keras.Sequential([l0,l1,l2])
model.compile(optimizer=tf.keras.optimizers.Adam(0.1),loss=tf.keras.losses.mean_squared_error)
history=model.fit(x,y,epochs=500,verbose=False)
print("Finished training the model")
```

```python
plt.plot(history.history["loss"])
```
